```
import pandas as pd 
df = pd.read_csv('2012_E_Results_US_Senate.csv')
print(df)
```

```
import matplotlib.pyplot as plt
_ = plt.hist(df['PERCVOTE2'], bins=20)
_ = plt.xlabel('oercent vote for Timothy M. Kaine (D)')
_ = plt.ylabel('number of Regions')
plt.show()
```

```
import seaborn as sns
sns.set()
_ = plt.hist(df['PERCVOTE2'], bins=20)
_ = plt.xlabel('oercent vote for Timothy M. Kaine (D)')
_ = plt.ylabel('number of Regions')
plt.show()
```

```
_ = sns.swarmplot(x='TOWNSHIP', y='WINPERCENT', data=df)
_ = plt.xlabel('state')
_ = plt.ylabel('percent of vote for Timothy M. Kaine (D)')
plt.show()
```

```
# Empirical cumulative distribution function (ECDF)
import numpy as np
x = np.sort(df['WINPERCENT'])
y = np.arange(1, len(x)+1) / len(x)
_ = plt.plot(x, y, marker='.', linestyle='none')
_ = plt.xlabel('percent of vote for Obama')
_ = plt.ylabel('ECDF')
plt.margins(0.04) # Keeps data off plot edges
plt.show()
```

```
# Chapter - 2
import numpy as np
s = np.mean(df['WINPERCENT'])
print(s)
```

```
s = np.median(df['WINPERCENT'])
print(s)
```

```
# Percentile
s = np.percentile(df['WINPERCENT'], [25,50,75])
print(s)
```

```
import matplotlib.pyplot as plt
import seaborn as sns
_ = sns.boxplot(x='PERCVOTE2', y='WINPERCENT', data=df)
_ = plt.xlabel('region')
_ = plt.ylabel('percent of vote for Obama')
plt.show()
```

```
# Varience
s = np.var(df['WINPERCENT'])
print(s)
```

```
# Standard Deviation
s = np.std(df['WINPERCENT'])
print(s)
```

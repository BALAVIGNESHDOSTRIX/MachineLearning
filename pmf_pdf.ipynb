```

# n =60, p = 0.1
samples = np.random.binomial(60, 0.5, size=2000)
print(samples)
```

```
def ecdf(data):
  """Compute ECDF for a one-dimensional array of measurements."""
  # Number of data points: n
  n = len(data)

  # x-data for the ECDF: x
  x = np.sort(data)

  # y-data for the ECDF: y
  y = np.arange(1, n+1) / n

  return x,y


import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()
x,y = ecdf(samples)
print(x,y)
_ = plt.plot(x, y, marker='.', linestyle='none')
plt.margins(0.02)
_ = plt.xlabel('number of successes')
_ = plt.ylabel('CDF')
plt.show()
```

```
samples = np.random.poisson(60,size=2000)
sns.set()
x,y = ecdf(samples)
print(x,y)
_ = plt.plot(x, y, marker='.', linestyle='none')
plt.margins(0.02)
_ = plt.xlabel('number of successes')
_ = plt.ylabel('CDF')
plt.show()
```

```
import pandas as pd 
df = pd.read_csv('2012_E_Results_US_Senate.csv')

def ecdf(data):
  """Compute ECDF for a one-dimensional array of measurements."""
  # Number of data points: n
  n = len(data)

  # x-data for the ECDF: x
  x = np.sort(data)

  # y-data for the ECDF: y
  y = np.arange(1, n+1) / n

  return x,y



import matplotlib.pyplot as plt
import numpy as np

mean = np.mean(df['WINPERCENT'])
samples = np.random.exponential(mean, size=10000)
x, y = ecdf(df['WINPERCENT'])
x_theor, y_theor = ecdf(samples)
_ = plt.plot(x_theor, y_theor)
_ = plt.plot(x, y, marker='.', linestyle='none')
_ = plt.xlabel('Number of Success')
_ = plt.ylabel('CDF')
plt.show()
```
